```python
import pandas as pd
import yfinance as yf
import datetime
import statsmodels.api as sm
import plotly
import plotly.graph_objects as go
import plotly.express as px
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
end = datetime.datetime(2021,9,15)
start = datetime.datetime(2021,8,1)
df = yf.download('AAPL', start=start, end=end)
df = df.reset_index()
df
```

```python
# fit model
model = SimpleExpSmoothing(df['Close'])
model_fit = model.fit()
# make prediction
yhat = model_fit.predict(0, len(df))
# create df
pred_df = pd.DataFrame()
pred_df['Date'] = df['Date']
pred_df['pred'] = yhat
pred_df['actual'] = df['Close']
pred_df
```

```python
model.forecast(steps=32)
```

```python
mod = sm.tsa.SARIMAX(df['Close'], order=(1, 0, 0), trend='c')
res = mod.fit()
print(res.forecast(steps=2))
```

```python
fig = px.line(pred_df, x=pred_df['Date'], y=pred_df['actual'], color_discrete_sequence=['grey'])
fig.add_traces(go.Scatter(x=pred_df['Date'], y=pred_df['pred'], name='EXP MA Prediction', line=dict(color='red')))
fig.show()
```

```python
lowess = sm.nonparametric.lowess
y_hat_lowess = lowess(pred_df['pred'].values, pred_df['actual'].values, frac=0.2)
fig = px.scatter(pred_df, x=pred_df['pred'], y=pred_df['actual'], color_discrete_sequence=['grey'])
fig.add_traces(go.Scatter(x=y_hat_lowess[:, 0], y=y_hat_lowess[:, 1], name='LOWESS pred',  line=dict(color='red')))
fig.show()
```
